Pattern:  
   
- 自底向上的全局更新模型（Bottom-up Global Update Pattern）  


- When: 题目要求解一个全局的、可能跨越左右子树的某种最值（如最大距离、最大路径、最长同值路径），但递归函数如果直接返回这个最值，信息无法在上下层之间连续传递。

- Key: “返回一个值，但在中途顺便更新另一个全局值”（Return one thing, update another globally）  


`````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        
        # Global variable to store the maximum path sum
        res = [root.val]

        # DFS function to compute the maximum path sum
        def dfs(root):
            # Base case: None
            if not root:
                return 0

            # return max sum without split
            leftMax = max(0, dfs(root.left))
            rightMax = max(0, dfs(root.right))

            # Compute max sum WITH split
            # Update the global maximum path sum if the current path sum is greater
            # Use res[0] instead of res to modify the value in the outer scope
            res[0] = max(res[0], root.val + leftMax + rightMax)

            return root.val + max(leftMax, rightMax)

        # Call the DFS function starting from the root
        dfs(root)

        # Return the maximum path sum found, the 1st element of res
        return res[0]
`````

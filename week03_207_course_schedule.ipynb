Topological sorting 拓扑排序

What：   
    - 所有点  ***按照先后顺序***  排成序列  
    - 必须是  ***有向无环图***  

How：   
    - 每次选   ***入度为0***   的点   
    - 再   ***删除***   这个点和它的出边   

Why：   
    - ***判断环***     
    （如果拓扑排序进行不下去了，说明图中有环）

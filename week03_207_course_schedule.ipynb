Topological sorting 拓扑排序

What：   
    - 所有点  ***按照先后顺序***  排成序列  
    - 必须是  ***有向无环图***  

How：   
    - 每次选   ***入度为0***   的点   
    - 再   ***删除***   这个点和它的出边   

Why：   
    - ***判断环***     
    （如果拓扑排序进行不下去了，说明图中有环）
    

Pattern 1：三色标记法 (Three-Color Marking)
这是图论遍历中用来检测环的工业级标准模式。在这段代码中，节点在任何时候都处于以下三种状态之一：

    - 状态 A（白色 - 未访问）： 该课程从未进入过 DFS，preMap[crs] 保持原样。

    - 状态 B（灰色 - 访问中）： 该课程存在于 visiting 集合中。这意味着它正在递归栈里，它的上游依赖还在被调查。一旦 DFS 再次碰撞到灰色节点，逻辑死循环确证，直接报错。

    - 状态 C（黑色 - 已安全）： 该课程已被彻底调查完毕且确认无环。在这里通过 preMap[crs] = [] 隐式表达。下次遇到直接放行。


Pattern 2：依赖图的拓扑排序 (Topological Sort)  

本题是拓扑排序的变体（环检测）。   

“任务依赖”、“编译顺序”、“模块打包”等有先后顺序要求的场景，标准的解题范式：

```构建邻接表 → 定义访问状态集 → DFS 深入递归 → 回溯与记忆化```

Business Problem:
 - 解决操作系统死锁检测、前端依赖打包（Webpack/Vite）

````python
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        
        # Map each course to its prerequisites
        # HashMap a adjList for each node O(1)
        # preMap = {crs: adjList of pre's}
        preMap = {i : [] for i in range(numCourses)}
        for crs, pre in prerequisites:
            preMap[crs].append(pre)


        # Store all courses along the current DFS path
        visiting = set()

        def dfs(crs):

            # Cycle detected
            if crs in visiting:
                return False

            # Base case: empty adjList
            if preMap[crs] == []:
                return True


            # Before visiting
            # Add cur crs to the visiting
            visiting.add(crs)

            # Recursively iterate all prerequisites
            for pre in preMap[crs]:
                if not dfs(pre):
                    return False
           
            # Backtrack: remove node from current path once fully explored
            visiting.remove(crs)
            
            # Set safety nodes/crs's adjList as empty
            # to use the basecase to avoid redundant check
            preMap[crs] = []
            return True

        # Global check to ensure all disconnected components are covered
        for c in range(numCourses):
            if not dfs(c):
                return False
        return True




````

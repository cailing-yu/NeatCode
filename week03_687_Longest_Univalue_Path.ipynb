# Problem Pattern: Longest Univalue Path

## 1. Key Thoughts
* **Bottom-Up Execution:** Parent nodes cannot resolve their optimal paths without first receiving computed states from their children.
* **Separation of Concerns:** You must independently compute and track two distinct metrics at every node:
    * *Extendable Path:* The maximum contiguous univalue path continuing down a single branch (`max(left, right) + 1`). Passed to the parent.
    * *Local Maximum:* The longest univalue path that peaks at the current node (`left + right`). Evaluated against the global maximum.
* **State Encapsulation:** Pure functions scale better. Eliminate side effects (like `nonlocal`) by passing required states upward explicitly via return values.

## 2. Core Pattern
* **Post-Order Depth-First Search (DFS):** Process both left and right subtrees completely before evaluating the parent root. 
* **Tree Dynamic Programming (Tree DP):** Solve subproblems at the leaf level and optimally propagate those states upward to establish the global maximum.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def longestUnivaluePath(self, root: Optional[TreeNode]) -> int:

        max_path = 0

        def dfs(root):

            nonlocal max_path

            if not root:
                return 0

            # Check left / right length seperately
            left_length = dfs(root.left)
            right_length = dfs(root.right)

            # Valid path
            left_path = 0
            right_path = 0

            # Check the same value of path
            if root.left is not None and root.left.val == root.val:
                left_path = left_length + 1
            if root.right is not None and root.right.val == root.val:
                right_path = right_length + 1


            # Check the max path
            max_path = max(max_path, left_path + right_path)

            return max(left_path, right_path)

        dfs(root)
        return max_path
```

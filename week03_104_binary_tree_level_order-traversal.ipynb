# BFS Level Order Traversal: Core Mechanics

## 1. Engineering Directives
* **State Tracking:** Execute mental dry runs. Variables must reflect actual state, not assumptions (e.g., measuring an unpopulated list).
* **Type Discipline:** Strictly differentiate between structural objects (`TreeNode`) and their primitive payloads (`int`).

## 2. Algorithmic Logic
* **Queue Initialization (`queue.append(root)`):** The `while` loop requires an initial coordinate to begin execution.
* **Optimal Structures (`deque`):** Use `collections.deque` for optimal $O(1)$ front-removals. Standard lists incur an $O(n)$ memory-shifting cost.
* **Type Contracts (`return []`):** Return an empty list for null inputs to fulfill the `List[List[int]]` signature and prevent downstream iteration crashes.
* **Boolean Evaluation (`while queue:`):** Populated collections naturally evaluate to `True`; empty ones evaluate to `False`.
* **Vessel vs. Payload:** Store the entire `TreeNode` object in the queue to maintain access to `.left` and `.right` pointers. Extract `.val` exclusively for the final result array.

```python
from collections import deque
from typing import Optional, List

class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        # Base Case: Fulfill return type contract
        if not root:
            return []
            
        # Initialization: Seed the starting coordinate
        queue = deque([root])
        res = []
        
        while queue:
            level = []
            size = len(queue) # Lock in the current level's node count
            
            for _ in range(size):
                node = queue.popleft()
                level.append(node.val) # Record payload
                
                # Queue structural vessels for the next level
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                    
            res.append(level)
            
        return res
```

# BFS Level Order Traversal: Complexity Analysis

## 1. Time Complexity: $O(N)$
* **$N$** = Total number of nodes in the binary tree.
* **Mechanism:** The algorithm visits every single node exactly one time. There are no redundant evaluations.
* **Operations:** Using `collections.deque` ensures that extracting a node (`popleft()`) and inserting its children (`append()`) are strictly $O(1)$ constant-time operations.
* **Conclusion:** $N$ nodes $\times$ $O(1)$ operations per node = strictly linear execution time, $O(N)$.

## 2. Space Complexity: $O(N)$
* **Peak Queue Memory:** The queue reaches its maximum memory footprint at the widest level of the tree. In a perfectly balanced tree, the bottom level contains roughly $\frac{N}{2}$ nodes. Dropping the constant multiplier results in $O(N)$ auxiliary space.
* **Result Array Memory:** The final return array (`res`) must store the payload (`.val`) of every single node in the tree, which inherently demands $O(N)$ space.
* **Conclusion:** Both the operational queue and the final data structure scale linearly with the input size, establishing a strict $O(N)$ space complexity.

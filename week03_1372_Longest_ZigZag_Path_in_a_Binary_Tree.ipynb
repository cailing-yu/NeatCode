# Algorithmic Analysis: Longest ZigZag Path in a Binary Tree

## 1. Engineering Weaknesses Identified
* **Tolerance for "Dead Code":** Leaving unused return statements (e.g., `return 0`) in a void context indicates trial-and-error coding rather than intentional, planned execution.
* **Incomplete Paradigm Shifts:** Mixing top-down state tracking (using a `nonlocal` variable) with bottom-up return values creates architectural friction and cognitive clutter.

## 2. Execution Strategies (The Fix)
* **Ruthless Refactoring:** Never stop at "it works." Delete any line that does not strictly contribute to the final result. If a function exclusively updates global state, it should return nothing (`None`).
* **Pre-Computation State Mapping:** Before writing code, explicitly define what data flows *down* the tree (passed as parameters) and what data flows *up* the tree (returned). Commit strictly to one paradigm.

## 3. Core Algorithmic Pattern
**Top-Down Depth-First Search (DFS) with State Passing.** Nodes do not just visit children; they pass down a customized "dossier" of state history (streak length) that dictates the child's next valid moves.

## 4. The 5-Step Thinking Process
1.  **Identify the Core Constraint:** A zigzag requires alternating directions. The decision at the current node depends entirely on the direction taken to arrive there.
2.  **Define the State (Parameters):** What must the node know from its parent? 
    * `left_path`: Streak length if the parent moved left.
    * `right_path`: Streak length if the parent moved right.
3.  **Define the Transitions:** How does state change upon moving?
    * *Move Left:* Extends a right-moving streak. (`left_path = right_path + 1`, reset `right_path = 0`).
    * *Move Right:* Extends a left-moving streak. (`right_path = left_path + 1`, reset `left_path = 0`).
4.  **Define Result Extraction:** A global tracker (`max_zigzag`) is most efficient here since the longest path can start and end anywhere in the tree.
5.  **Define the Base Case:** What terminates the recursion? Reaching a null node (`if not node: return`).

```python
# Defhttps://leetcode.com/problems/longest-zigzag-path-in-a-binary-tree/$0inition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def longestZigZag(self, root: Optional[TreeNode]) -> int:

        # Global variable to store the result
        max_zigzag = 0

        # dfs function to find th zigzag
        # 3 args: root, l, r
        def dfs(node, left_path, right_path):

            
            nonlocal max_zigzag

            # Base case:
            if not node: return 0

            # Recyrsuve transitions
            # # Moving left extends a path that previously moved right
            dfs(node.left, right_path + 1, 0)

            # Move right
            dfs(node.right, 0, left_path + 1)

            # Updating global state
            max_zigzag = max(max_zigzag, left_path, right_path)


        dfs(root, 0, 0)
        return max_zigzag


```

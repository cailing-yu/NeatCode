````python
# Pattern

class Template:
    def solveTreeProblem(self, root: Optional[TreeNode]) -> int:
        if not root: 
            return 0
        
        # Core: Create a global scoreboard using a mutable single-element list
        global_record = [INITIAL_WORST_VALUE]  # Usually 0 or float('-inf')
        
        def bottom_up_worker(node):
            if not node:
                return BOUNDARY_VALUE  # Usually 0 or -1
                
            # 1. Recursively request the core metric from subtrees
            left_metric = bottom_up_worker(node.left)
            right_metric = bottom_up_worker(node.right)
            
            # 2. [Core Side-Effect] Combine subtree metrics to calculate the current 
            #    cross-node value, and challenge the record at index [0]
            current_cross_value = combine_logic(left_metric, right_metric, node.val)
            global_record[0] = max(global_record[0], current_cross_value)
            
            # 3. Stay loyal to the recursive mission: return a valid, unbranched, 
            #    single-sided continuous metric to the parent node
            return report_to_parent_logic(left_metric, right_metric, node.val)
            
        # Trigger the pipeline
        bottom_up_worker(root)
        
        # Extract the ultimate winner from the list
        return global_record[0]
````

Key:
    - Global variable


- When using ```longest_diameter = 0```, it is an immutable integer. To modify it inside a nested function, you must declare it as ```nonlocal```.

- When using ```longest_diameter = [0]```, it is a list (a pointer to a fixed memory slot). Writing ```longest_diameter[0] = ...``` changes the content inside the slot without moving the slot itself.

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:

        # Set a global variable to store the result
        longest_diameter = 0

        # Create a helper function to find the longest diameter
        def height(root):

            nonlocal longest_diameter

            # Base case: None value
            if not root:
                return 0

            # Calculate the height of left subtree, right subtree
            left_height = height(root.left)
            right_height = height(root.right)

            # Calculate current diameter
            cur_diameter = left_height + right_height

            # Longest diameter
            longest_diameter = max(cur_diameter, longest_diameter)

            return 1 + max(left_height, right_height)

        # Call the function and return the result
        height(root)
        return longest_diameter
        
````

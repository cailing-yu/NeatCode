````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:

        # Base Case:
        if not root: return 0

        # Calculate the # of left/right nodes
        left_nodes = self.maxDepth(root.left)
        right_nodes = self.maxDepth(root.right)

        # Return the deepest path between the two, PLUS 1 for the current node itself.
        return 1 + max(left_nodes, right_nodes)
````

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:

        # Base Case:
        if not root: return 0

        # Return the deepest path between the two, PLUS 1 for the current node itself.
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
````

# Algorithm: Maximum Depth of Binary Tree

## 1. Key Thoughts
* **Pure Function Design:** Compute and return the metric directly up the call stack. External state tracking is unnecessary.
* **Systematic Delegation:** The depth of any given tree is exactly `1` (accounting for the current node) plus the maximum depth of its subtrees.
* **Absolute Base Case:** An empty node (`not root`) fundamentally contributes `0` to the depth.



## 2. Pitfalls
* **State Contamination:** Relying on global or `nonlocal` variables when a stateless synthesis works creates fragile code.
* **Redundant Wrappers:** Building a helper function when the primary method signature already dictates the exact input/output needed for recursion.
* **Infinite Recursion:** Calling the method on the current `root` instead of advancing the state via `root.left` and `root.right`.
* **Omission of Current State:** Forgetting to add `1` at the return step, failing to account for the current node.



## 3. Pattern
* **Bottom-Up Recursion (Post-order Traversal):** The execution traverses to the leaf nodes, hits the base cases, and resolves the calculations as the stack unwinds back to the top.
* **Stateless Synthesis:** The answer is constructed entirely by comparing return values at each level (`max(left, right)`).
* **Time Complexity:** `O(N)` — Every node is visited exactly once.
* **Space Complexity:** `O(H)` — Where `H` is the height of the tree, representing the memory used by the system call stack (Worst case `O(N)` for skewed trees, `O(log N)` for balanced trees).

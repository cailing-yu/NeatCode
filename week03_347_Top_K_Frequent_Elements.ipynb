- Problem: 347. Top K Frequent Elements  
- Topics: Heap(Priority Queue), Hash Table, Counting, Bucket Sort

- Business:
    - 从海量数据中，动态或静态地筛选出最具代表性、发生频率最高的核心特征。
        - 搜索引擎与社交媒体：热搜与高频词汇统计: 微博热搜、百度热度榜、Google Trend
        - 电商与内容平台：热门商品与爆款推荐: 亚马逊的“Best Sellers”榜单、抖音的“热门视频”、Spotify 的“每周最热歌曲”
        - 网络安全与系统运维：异常流量与攻击检测（Heavy Hitters 问题）: 防火墙防 DDoS 攻击、服务器日志分析。
        - 金融风控：高频交易与套现识别: 银行信用卡防刷、反洗钱。

- Patterns:

    - 核心步骤统一分为两步：

        - 计数（Frequency Map）
            - 使用 Hash Map 统计每个元素出现的频率。时间复杂度 O(N)，空间复杂度 O(N)。

        - 筛选（Top K Selection）：
                如何从 Hash Map 中找出前 K 个高频元素。


    - Pattern 1: 最小堆 Min_Heap / Priority Queue (注意：求最高频用最小堆，求最低频用最大堆)
        - minheap that’s kept at size k,
        - if its bigger than k pop the min,
        - by the end it should be left with k largest


    - Pattern 2: 桶排序（Bucket Sort）—— 极致的时间复杂度 O(N)

Pattern 1：最小堆（Min-Heap / Priority Queue）

- 核心逻辑：

    - 遍历 Hash Map。

    - 维护一个大小为 K 的最小堆（注意：求最高频用最小堆，求最低频用最大堆）。

    - 当堆的尺寸超过 K 时，弹出堆顶元素（即当前堆中频率最低的元素）。

    - 遍历结束后，堆中剩下的 K 个元素就是高频的前 K 个。

- 复杂度分析：

    - 时间复杂度：O(NlogK)。
        遍历 Map 需要 O(N)，每次堆操作需要 O(logK)。  
        因为 K≤N，当 K 远小于 N 时，这个效率非常高。

    - 空间复杂度：O(N+K)。
        Hash Map 空间 O(N)，堆空间 O(K)。

- Key:
    - ```Counter()```   Dict subclass for counting hashable objects.  

    - ```Counter.most_common()	```  Return a list of the n most common elements and their counts.
        - find the k most frequent elements
        - returning them as a list of (element, count) tuples
    
    - ```[x for x, _ in cnt.most_common(k)] ```extracts just the elements, discarding the counts.

````python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:

        # Count Frequencies
        freq = Counter(nums)

        # Get K most elements, returns frequencies(elem, count)
        mostCommonElems = freq.most_common(k)

        # Extract elems, return result
        res = [element for element, count in mostCommonElems]

        return res
````

Time Complexity: O(n log k) where n is the length of the input array. We traverse the array once (O(n)) and perform heap operations for each unique element, where each operation takes O(log k).

Space Complexity: O(n) for the hash table to store frequencies, plus O(k) for the heap.

Pattern 2：桶排序（Bucket Sort）—— 极致的时间复杂度 O(N)

当要求时间复杂度必须达到 O(N) 时，桶排序是首选。

- 核心逻辑：

    - 利用“频率”作为数组的下标（桶的编号）。因为一个元素出现的最高频率不可能超过原数组的长度 N。

    - 创建一个大小为 N+1 的数组（桶），数组的每个位置是一个列表，用来存放对应频率的元素。

    - 逆序（从频率最高的桶开始）收集桶里的元素，直到凑齐 K 个。

- 复杂度分析：

    - 时间复杂度：O(N)。无论是建表、入桶还是逆序收集，都只需要线性时间。

    - 空间复杂度：O(N)。需要额外的桶空间来存储元素。

````python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:

        # Count Frequencies
        count = {} # Hash map occurence of each val
        freq = [[] for i in range(len(nums) + 1)]
        
        # Count how many times each val occurs
        for n in nums:
            count[n] = 1 + count.get(n, 0)

        # Iterate every val we counted, Returns {key: val} pairs; count would be the index
        for n, c in count.items():
            # Val n occurs c times
            freq[c].append(n)

        # Results
        res = []

        # Iterate freq arry in DESC order
        # range(Start: last index, Stop: index 0, DESEC order)
        for i in range(len(freq) - 1, 0, -1):
            for n in freq[i]:
                res.append(n)

                # Extract K vals
                if len(res) == k:
                    return res
````

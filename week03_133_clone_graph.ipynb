````python
from typing import Optional

class Solution:
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']:
        # 卫语句：处理极端边界（空图）
        if not node:
            return None

        # 核心工具：新旧节点映射表（防死循环，建立连接）
        clones = {}

        # 核心驱动：DFS 递归函数
        def dfs(curr_node):
            # 1. 审查：如果已经克隆过，直接返回克隆体
            if curr_node in clones:
                return clones[curr_node]

            # 2. 创建与登记：创建新节点对象（自动包含空的 neighbors 列表）
            copy_node = Node(curr_node.val)
            clones[curr_node] = copy_node

            # 3. 组装：遍历原邻居，递归克隆它们，并装入新节点的 neighbors 列表中
            for neighbor in curr_node.neighbors:
                # 隐式系统栈在这里工作：暂停当前，深入下一层，返回后 append
                copy_node.neighbors.append(dfs(neighbor))

            # 返回组装完毕的新节点
            return copy_node

        # 启动引擎
        return dfs(node)
````

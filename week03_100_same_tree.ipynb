```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:

        # BaseCase1: Both empty -> True
        if not p and not q: return True

        # Mismatch detection1: One is empty -> False
        if not p or not q: return False

        # Mismatch detection2:  Only one node is null or the values don't match -> False
        if p.val != q.val: return False

        # Recursive check: If current val is same, check left and right subtrees
        return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:

        # BaseCase1: Both empty -> True
        # if not p and not q: return True
        if p is None and q is None: return True
        

        # Mismatch detection1: One is empty -> False
        # Mismatch detection2:  Only one node is null or the values don't match -> False
        # if not p or not q or p.val != q.val: return False
        if p is None or q is None or p.val != q.val: return False

        
        # Recursive check: If current val is same, check left and right subtrees
        return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
```

### 1. Key Thoughts (Mental Models)

* **Strict Call Stack:** Execution within a frame is strictly top-down. A `return` statement immediately terminates it.
* **MECE Base Cases:** Exhaustively map all physical input states (Mutually Exclusive, Collectively Exhaustive): 
    1. Both empty. 
    2. Exactly one empty. 
    3. Both exist, but values differ.
* **Divide and Conquer:** Validate the local node's state, then strictly delegate identical validation to subtrees.

### 2. Pitfalls (Cognitive Blind Spots)

* **Misinterpreting `self`:** `self` refers to the executing class instance (`Solution`), not the data entity (`TreeNode`). You are invoking class behavior.
* **Boolean Short-Circuiting:** `return A and B` does not mean "return True". It evaluates dynamically. If `A` is `False`, `B` is entirely ignored to optimize runtime.
* **Flattening Recursion:** Recursion is not a linear script. It is a dynamically expanding stack of independent frames that must resolve bottom-up.

### 3. Pattern (Standard Tree Recursion & Syntax Alternatives)

*Note: While Python allows implicit checks (`not p`), explicit identity checks (`p is None`) are often preferred in strict engineering environments to prevent edge-case bugs with objects evaluating to `False`.*

**1. Success Base Case (The Bottom):** Reached the end of a valid branch safely.
* *Implicit:* `if not p and not q: return True`
* *Explicit:* `if p is None and q is None: return True`

**2. Failure Base Case (The Interceptor):** Structural asymmetry or value mismatch at the current level.
* *Implicit:* `if not p or not q or p.val != q.val: return False`
* *Explicit:* `if p is None or q is None or p.val != q.val: return False`

**3. Recursive Step (The Delegation):** If the current node passes checks, delegate to children and combine logic.
* `return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)`

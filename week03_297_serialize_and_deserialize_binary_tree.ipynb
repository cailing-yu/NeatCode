The Architectural Symmetry:

Notice how the pattern perfectly mirrors itself across both operations:

1. Serialization Pattern:

    - Action: Read current node → Recursively process Left → Recursively process Right.

    - Base Case: If None, append 'N'.

2. Deserialization Pattern:

    - Action: Consume current array token to build node → Recursively assign to .left → Recursively assign to .right.

    - Base Case: If token is 'N', advance pointer and return None.

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """
        Encodes a tree to a single string.
        将二叉树扁平化为一维字符串（前序遍历 DFS）
        """
        res = []

        def dfs(node):
            if not node:
                res.append('N')  # 遇到空节点，用 'N' 占位并回溯
                return
            res.append(str(node.val))  # 记录当前根节点的值
            dfs(node.left)             # 递归进入左子树
            dfs(node.right)            # 递归进入右子树

        dfs(root)
        return ','.join(res)  # 用逗号连接成最终的字符串


    def deserialize(self, data):
        """
        Decodes your encoded data to tree.
        将一维字符串还原为标准的二叉树结构
        """
        vals = data.split(',')  # 将字符串切割回列表形式
        self.i = 0  # 核心：维护一个全局/类维度的指针，记录当前处理到列表的哪个位置

        def dfs():
            # Base Case: 如果当前指针指向的是 'N'，说明是个空节点
            if vals[self.i] == 'N':
                self.i += 1  # 指针向后移动一位，消费掉这个 'N'
                return None  # 向上一层返回 None

            # 核心逻辑：按前序遍历的逆矩阵操作，重建当前节点
            node = TreeNode(int(vals[self.i]))  # 实例化当前根节点
            self.i += 1  # 指针向后移动一位，消费掉当前数字
            
            # 顺次消费后续的 Token，通过递归将子树挂载到对应的左右指针上
            node.left = dfs()   # 左层递归结束后，self.i 会被精准推移到右子树的起点
            node.right = dfs()  # 右层递归消费完毕后，返回完整的局部树结构
            
            return node

        return dfs()  # 启动递归，并返回整棵树的根节点

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
````
